Matrix Operations Using Numpy

```python
X = np.random.randn(2,3)    # create a 2 x 3 random matrix

print(X)

print(X.T)             # transpose operation X^T

y = np.random.randn(3) # random vector
print("value of y")
print(y)
print("matrix-vector")
print(X.dot(y))        # matrix-vector multiplication  X * y
print("matrix-matrix")
print(X.dot(X.T))      # matrix-matrix multiplication  X * X^T
print(X.T.dot(X))      # matrix-matrix multiplication  X^T * X
```

### Aggregating Dataframes

groupby command creates groups that can be aggregated

```python
# show the mean of all columns where data has been grouped by rank
salaries.groupby('discipline').mean()  # group by => categorical
```

```python
# compute median statistics for one column
salaries.groupby('rank')['salary'].median()
```

Filtering data based on column values

```python
salaries[salaries['salary'] > 1000]
```

```python
# logical AND
salaries[(salaries['rank'] == 'Prof') & (salaries['discipline'] == 'A')]
```

```python
# logical OR
salaries[(salaries['rank'] == 'Prof') | (salaries['discipline'] == 'A')]
```

```python
# salaries pivot table example
salaries.pivot_table(index='discipline', columns='rank',
                    aggfunc={'salary':'count'})
```

```python
rng = np.random.RandomState(0)
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data1': range(6),
                   'data2': rng.randint(0, 10, 6)},
                   columns = ['key', 'data1', 'data2'])
df
```

```python
# another example of aggregation
df.groupby('key').aggregate({'data1': 'min',
                             'data2': 'max'})
```

Filtering groups

```python
def filter_func(x):
    return x['data2'].std() > 4
```

```python
# display those keys where filter_func is true i.e. those keys where data2.std() is greater than 4
df.groupby('key').filter(filter_func)
```

```python
# count of distinct values
df['key'].value_counts()
```

```python
# sort values of a column
df['data2'].sort_values(ascending=False)
```

The apply() method lets you apply an arbitrary function to the group results.

```python
def calc_sum(x):
  return x.sum()

data = {
  "x": [50, 40, 30],
  "y": [300, 1112, 42]
}


df1 = pd.DataFrame(data)

df1

# axis = 1 applies the function by index i.e. row-wise, axis = 0 columnwise
df1.apply(calc_sum, axis=1)
```

```python
# axis = 1 applies the function by rows
df1.apply(calc_sum, axis=1)
```

## Python Time Series Data Analysis

Let's look at a time series example of NO<sub>2</sub> levels in various places across time.

```python
air_quality = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/air_quality_no2_long.csv", parse_dates=["date.utc"])
```

```python
air_quality['country'] = air_quality['country'].apply(str)
```

```python
air_quality.info()
```

```python
air_quality.head()
```

```python
air_quality = air_quality.rename(columns={"date.utc": "datetime"})
```

```python
air_quality
```

```python
air_quality.info()
```

```python
air_quality['datetime'].dt.month
```

```python
air_quality['datetime'].max() - air_quality['datetime'].min()
```

```python
air_quality["month"] = air_quality["datetime"].dt.month
air_quality["day"] = air_quality["datetime"].dt.day
```

```python
air_quality.info()
```

```python
air_quality
```

```python
# find means grouped by 2 quantities - day of week first and then location
air_quality.groupby([air_quality["datetime"].dt.weekday, "location"])["value"].mean()
```

```python
import matplotlib.pyplot as plt # plt is an alias for

fig, axs = plt.subplots(figsize=(12, 4))
air_quality.groupby(air_quality["datetime"].dt.hour)["value"].mean().plot(kind='bar', rot=0, ax=axs)
plt.xlabel("Hour of the day")  # custom x label using matplotlib
plt.ylabel("$NO_2 (Âµg/m^3)$")
plt.title("My first plot")
```
